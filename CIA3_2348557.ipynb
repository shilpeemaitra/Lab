```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.linear_model import LogisticRegression
X = data.data[:, 2:4]  # Using petal length and petal width as features
y = data.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a logistic regression model with only two features
model = LogisticRegression(multi_class='ovr', solver='liblinear')
model.fit(X_train, y_train)

# Visualize the decision boundary on a scatter plot (petal length vs. petal width)
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.6)

# Scatter plot of the actual data points
scatter = plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Logistic Regression Decision Boundary for Iris Species Classification')
plt.legend(handles=scatter.legend_elements()[0], labels=data.target_names)
plt.show()

```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score

# Make predictions on the testing data
y_pred = model.predict(X_test)

# Calculate the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Calculate precision (Setosa class)
precision_setosa = precision_score(y_test == 0, y_pred == 0)

# Calculate precision (Versicolor class)
precision_versicolor = precision_score(y_test == 1, y_pred == 1)

# Calculate precision (Virginica class)
precision_virginica = precision_score(y_test == 2, y_pred == 2)

print("Confusion Matrix:")
print(conf_matrix)
print("\nAccuracy:", accuracy)
print("Precision (Setosa):", precision_setosa)
print("Precision (Versicolor):", precision_versicolor)
print("Precision (Virginica):", precision_virginica)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier


X = data.data
y = data.target

# Binarize the target variable (one-vs-all approach)
y = label_binarize(y, classes=[0, 1, 2])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a logistic regression model (multi-class)
model = OneVsRestClassifier(LogisticRegression(solver='liblinear'))
model.fit(X_train, y_train)

# Predict probabilities for each class
y_score = model.predict_proba(X_test)

# Compute ROC curve and AUC for each class
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(3):  # Three classes (Setosa, Versicolor, Virginica)
    fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Compute micro-average ROC curve and AUC
fpr["micro"], tpr["micro"], _ = roc_curve(y_test.ravel(), y_score.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

# Plot ROC curve for each class and micro-average
plt.figure(figsize=(10, 6))
plt.plot(fpr[0], tpr[0], color='darkorange', lw=2, label='ROC curve (Setosa) (AUC = {:.2f})'.format(roc_auc[0]))
plt.plot(fpr[1], tpr[1], color='green', lw=2, label='ROC curve (Versicolor) (AUC = {:.2f})'.format(roc_auc[1]))
plt.plot(fpr[2], tpr[2], color='blue', lw=2, label='ROC curve (Virginica) (AUC = {:.2f})'.format(roc_auc[2]))

plt.plot(fpr["micro"], tpr["micro"], color='red', lw=2, label='Micro-average ROC curve (AUC = {:.2f})'.format(roc_auc["micro"]))

plt.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve for Iris Species Classification')
plt.legend(loc="lower right")
plt.show()

# Display AUC scores
print("AUC (Setosa):", roc_auc[0])
print("AUC (Versicolor):", roc_auc[1])
print("AUC (Virginica):", roc_auc[2])
print("Micro-average AUC:", roc_auc["micro"])

# it plots the True Positive Rate (TPR) against the False Positive Rate (FPR)
```
